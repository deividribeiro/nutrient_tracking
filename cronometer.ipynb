```python
# Assuming your DataFrame is called 'df' with date index and category columns
# Transpose the DataFrame to have categories on x-axis and dates on y-axis

df_transposed = differences[fields_to_plot]

# Create a custom colormap
cmap = sns.diverging_palette(10,130, as_cmap=True)

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_transposed, cmap=cmap, center=0, 
            cbar_kws={'label': 'Value'})

# Customize the plot
plt.title('Heatmap of Categories over Time')
plt.xlabel('Categories')
plt.ylabel('Date')

# Rotate x-axis labels if there are many categories
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.savefig("figures/nutrients_raw.png")
plt.show()
```

```python
df_transposed_clean = differences[fields_to_plot]

# Define the smallest value for the logarithmic scale (to avoid issues with 0)
linthresh = 1e-2  # Threshold below which the scale behaves linearly
vmin = df_transposed_clean.min().min()
vmax = df_transposed_clean.max().max()

# Create a symmetric logarithmic norm
norm = SymLogNorm(linthresh=linthresh, vmin=vmin, vmax=vmax)

# Create the heatmap
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(
    df_transposed_clean,
    cmap=cmap,
    norm=norm,
    cbar_kws={'label': 'Value (Symmetric Log Scale)'}
)

# Customize the colorbar ticks
colorbar = heatmap.collections[0].colorbar
colorbar.set_ticks([-vmax, -1, -linthresh, 0, linthresh, 1, vmax])
colorbar.set_ticklabels([f'{x:.1e}' for x in colorbar.get_ticks()])

# Customize the plot
plt.title('Heatmap of Categories over Time (Symmetric Log Scale)')
plt.xlabel('Categories')
plt.ylabel('Date')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("figures/nutrients_log.png")
plt.show()
```
